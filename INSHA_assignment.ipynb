```python
# Write a Pandas program to calculate the mean score for each different student in DataFrame.
x["score"].mean()
```

```python
# Write a Pandas program to append a new row 'k' to data frame with given values for each column.
# y = pd.DataFrame({"name": 'Froggy',"score": 99.9,"attempts":2,"qualify":"yes"},index=['k'])
# x.append(y)
x.loc["k"]=["Froggy",99.9,2,"yes"]
x
```

```python
# Write a Pandas program to sort the DataFrame first by 'name' in descending order, then by 'score' in ascending order
# print("This is sorted by name in descending order\n")
# print(x.sort_values("name")[::-1])
# print("This is sorted by score in ascending order\n")
# print(x.sort_values("score")[::1])

x.sort_values(by=["name","score"],ascending=[False,True])
```

```python
# Write a Pandas program to replace the 'qualify' column contains the values 'yes' and 'no' with True and False

x.loc[x["qualify"]=="yes","qualify"] = True
x.loc[x["qualify"]=="no","qualify"] = False
x
```
